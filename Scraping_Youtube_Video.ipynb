## Scrapping Data
Melakukan scrapping data terkait Clash of Champion by Ruangguru. hasil scrapping berupa komentar terkait video Clash of Champion yang tersedia di Youtube

```python
# import library
import googleapiclient.discovery
import pandas as pd
```

```python
# api key
API_KEY = "KEY"
```

```python
# variable parameters
api_service_name = "youtube"
api_version = "v3"
DEVELOPER_KEY = API_KEY
```

```python
# create client api
youtube = googleapiclient.discovery.build(
    api_service_name, api_version, developerKey=DEVELOPER_KEY)
```

```python
def get_comments(client, videoId):
  youtube = client
  comments=[]
  nextPageToken = None

  while True:
    req = youtube.commentThreads().list(
        part='snippet',
        videoId=videoId,
        maxResults=100,
        pageToken=nextPageToken
    )

    res = req.execute()
    for item in res['items']:
      comment = item['snippet']['topLevelComment']['snippet']
      comments.append([
          comment['authorDisplayName'],
          comment['publishedAt'],
          comment['likeCount'],
          comment['textOriginal']
      ])

    nextPageToken = res.get('nextPageToken')
    if not nextPageToken:
      break

  return pd.DataFrame(comments, columns=['author', 'dates', 'likes', 'Comments'])
```

```python
# Melakukan Scraping data
id = "dY_3acmFFRQ"
df = get_comments(youtube, id)
```
